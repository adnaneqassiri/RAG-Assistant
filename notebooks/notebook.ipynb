```python
test_cases = [
    # (query, expected_retrieval)
    
    # — General knowledge → FALSE
    ("What is the capital of Canada?",              False),
    ("What is machine learning?",                   False),
    ("What is JDBC?",                               False),
    ("Explain TCP vs UDP",                          False),
    ("What is a neural network?",                   False),
    ("What is RAG?",                                False),
    ("Explain ESRGAN",                              False),
    ("What is SISR?",                               False),
    
    # — Explicit doc references → TRUE
    ("How is ESRGAN trained? from the SISR document", True),
    ("According to the uploaded file, what is the main algorithm?", True),
    ("Based on this document, explain the architecture", True),
    ("What does the course say about attention mechanisms?", True),
    
    # — Structural patterns → TRUE (rule_override)
    ("Explain question 3 of TD2",                   True),
    ("What is the answer to exercise 1?",           True),
    ("Solve question 5 from TD3",                   True),
    ("What does chapter 2 say about transformers?", True),
    ("Summarize section 3",                         True),
    ("See figure 4 in the document",                True),
]

# Run & evaluate
results = []
correct = 0

for query, expected in test_cases:
    result = retrieval_router(query)
    actual = result["retrieval_needed"]
    ok = actual == expected
    if ok:
        correct += 1
    results.append({
        "query": query,
        "expected": expected,
        "actual": actual,
        "source": result["decision_source"],
        "pass": "✅" if ok else "❌"
    })

# Print report
print(f"\n{'='*60}")
print(f"SCORE: {correct}/{len(test_cases)} ({100*correct//len(test_cases)}%)")
print(f"{'='*60}\n")

for r in results:
    print(f"{r['pass']} [{r['source']:<15}] {r['query'][:50]}")
    if not r["pass"] == "✅":
        print(f"     Expected: {r['expected']} | Got: {r['actual']}")
```
